```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([7,3])
b = np.array([3,7])
ax = plt.axes()

#a
ax.arrow(0,0,a[0],a[1], head_width = 0.2)
plt.annotate(f"a({a[0] , a[1]})" , xy=(a[0]+0.5,a[1]))

#b

ax.arrow(0,0,b[0],b[1], head_width = 0.2)
plt.annotate(f"b({b[0] , b[1]})" , xy=(b[0]+0.5 , b[1]))
plt.xlim(0,10)
plt.ylim(0,10)

cos_sin = np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
print(cos_sin)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
a = np.array([7,3])
b = np.array([3,7])
print(a.shape)
c = a.reshape(1,-1)
print(c.shape)
cos_sin = cosine_similarity(a.reshape(1,-1),b.reshape(1,-1))
cos_sin
```

```python
doc_trump = "Mr. Trump became president after winning the political election. Though he lost the support of some republican friends, Trump is friends with President Putin"

doc_election = "President Trump says Putin had no political interference is the election outcome. He says it was a witchhunt by political parties. He claimed President Putin is a friend who had nothing to do with the election"

doc_putin = "Post elections, Vladimir Putin became President of Russia. President Putin had served as the Prime Minister earlier in his political career"

doc_learnbay = 'We are here to learn data science'

documents = [doc_trump, doc_election, doc_putin, doc_learnbay]
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd

cv = CountVectorizer()
tf = TfidfVectorizer()

sm = cv.fit_transform(documents)
sm_2 = tf.fit_transform(documents)
```

```python
import pandas as pd
df = pd.DataFrame(sm.todense() , columns = cv.get_feature_names() , index = ['doc_trump', 'doc_election', 'doc_putin', 'doc_learnbay'])
df
```

```python
df1 = pd.DataFrame(sm_2.todense() , columns = cv.get_feature_names() , index = ['doc_trump', 'doc_election', 'doc_putin', 'doc_learnbay'])
df1
```

```python
cosine_similarity(df1,df1) #tfid
```

```python
cosine_similarity(df,df) #countvectorise
```
