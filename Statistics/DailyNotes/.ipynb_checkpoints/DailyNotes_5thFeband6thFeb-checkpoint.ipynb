```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

#1000 simulatos of die roll
n = 10

avg = []
#In each simulation there is 1 trial more than the prev. simulation
for i in range(2,n):
    a = np.random.randint(1,7,i)
    avg.append(np.average(a))

#Fun. that will plot the hist. where current is the latest fig
def clt(current):
    plt.cla()
    if current == 10:
        a.event_source.stop()
    plt.hist(avg[0:current])
    
    plt.gca().set_title('Exp. value of die rolls')
    plt.gca().set_xlabel('Avg. from die roll')
    plt.gca().set_ylabel('Frequency')
    
    plt.annotate('Die roll = {}'.format(current),[3,27])
    
fig = plt.figure()
a = animation.FuncAnimation(fig,clt,interval=5)
plt.show()
```
