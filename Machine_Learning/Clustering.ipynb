```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r'DataSets\Wholesale_customers_data.csv')
print(df.head())

df['Fresh'].plot.line(title='info')
df['Milk'].plot.line(title='info')
```

```python
from sklearn.preprocessing import normalize
normal_df = pd.DataFrame(normalize(df), columns = df.columns)
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(normal_df, method='ward'))
plt.axhline(y=5, color='r', linestyle='--')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')  
cluster.fit_predict(normal_df)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(normal_df['Milk'], normal_df['Grocery'], c=cluster.labels_) 
plt.show()
```
