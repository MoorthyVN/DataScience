```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10,10)
y = 1/(1 + np.exp(-x))
```

```python
plt.plot(x,y)
plt.show()
```

```python
h_theta1 = -(np.log(y))
```

```python
h_theta2 = -(np.log(1-y))
```

```python
plt.plot(y,h_theta1)
plt.show()
```

```python
plt.plot(y,h_theta2)
plt.show()
```
